# Python OOP

### 1

OOP (Object oriented Programing) adalah suatu metode pemrograman yang berorientasi objek, atau secara sederhananya, OOP adalah paradigma pemrograman yang merepresentasikan objek seperti objek di sekitar kita dalam bentuk kode program

```python
class Topi(object):
    
    def __init__(self, input_milik, input_warna, input_jenis):
        self.milik = input_milik
        self.warna = input_warna
        self.jenis = input_jenis

    def panas(self):
        print("pakai topi ", self.milik)

    def info(self):
        print(f"milik: {self.milik}, warna: {self.warna}, jenis: {self.jenis}")


topi1 = Topi("Dina", "putih", "topi pantai")

topi1.info()
topi1.panas()
```

```python
from abc import abstractmethod, ABC
```

```python
class Komik(ABC):
    
    @abstractmethod
    def asal(self):
        pass
    
class Manhwa(Komik):
    def asal(self):
        print("Korea")
        
class Manga(Komik):
    def asal(self):
        print("Jepang")
```

```python
komik1= Manhwa()
komik2= Manga()

komik1.asal()
komik2.asal()
```

# Machine Learning

### 1

Supervised learning adalah algoritma machine learning yang proses pembelajarannya di bawah pengawasan guru atau supervisor. Contohnya: meemprediksi kepuasan pelanggan, memprediksi harga real estate

Unsupervised Learning adalah algoritma machine learning yang proses pembelajarannya tidak perlu pengawasan. Contohnya: proses segmentasi pelanggan, sistem rekomendasi pada ecommerce
